```python
import numpy as np
import pandas as pd
import glob
from sklearn.model_selection import train_test_split
import os
```

```python
data = pd.read_csv('../input/rsna-2023-abdominal-trauma-detection/train.csv')
```

```python
data.columns
```

```python
def get_image_file(row):
    patient_id = row['patient_id']
    image_folder = f'../input/rsna-2023-abdominal-trauma-detection/train_images/{patient_id}'
    image_files = glob.glob(os.path.join(image_folder,'**','*.dcm'),recursive=True)
    return image_files
```

```python
data['image_files'] = data.apply(get_image_file, axis=1)

# Save the updated DataFrame back to the CSV file
#df.to_csv(csv_file_path, index=False)
```
